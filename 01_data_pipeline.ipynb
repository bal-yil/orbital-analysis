```python
import requests
import pandas as pd
from sgp4.api import Satrec, WGS72

# Pull active satellites TLE catalog from Celestrak
url = "https://celestrak.org/SOCRATES/query.php"
url = "https://celestrak.org/pub/TLE/catalog.txt"

response = requests.get(url)
lines = response.text.strip().splitlines()

print(f"Total lines pulled: {len(lines)}")
print("\nFirst 9 lines (first 3 TLE entries):")
for line in lines[:9]:
    print(line)

```
